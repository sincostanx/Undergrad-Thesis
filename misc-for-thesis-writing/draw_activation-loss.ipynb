```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Sigmoid(x):
  return 1/(1+np.exp(-x))
def LeakyReLU(x, a=0.1):
  return np.maximum(np.zeros(x.shape),x) + a * np.minimum(np.zeros(x.shape),x)
```

```python
plt.rcParams['figure.figsize'] = [20, 5]
x = np.arange(-10., 10., 0.2)

plt.subplot(1, 3, 1)
plt.xlabel("Value of input (x)")
plt.title("Sigmoid function")
plt.plot(x, Sigmoid(x))

plt.subplot(1, 3, 2)
plt.xlabel("Value of input (x)")
plt.title("Tanh function")
plt.plot(x, np.tanh(x))

plt.subplot(1, 3, 3)
plt.xlabel("Value of input (x)")
plt.title("LeakyReLU function (a = 0.1)")
plt.plot(x, LeakyReLU(x))
plt.savefig("activation-func.pdf")
```

```python
def CrossEntropy(yHat, y):
    return -y*np.log(yHat) + (1 - y)*np.log(1 - yHat)
def L2(yHat, y):
    return (yHat - y)**2
def L1(yHat, y):
    return np.abs(yHat - y)
```

```python
plt.rcParams['figure.figsize'] = [20, 5]
x = np.arange(-10., 10., 0.2)

plt.subplot(1, 3, 1)
plt.xlabel("Value of input (x)")
plt.title("L1 loss, ground truth (y_hat) = 0")
plt.plot(x, L1(x, np.zeros(x.shape)))

plt.subplot(1, 3, 2)
plt.xlabel("Value of input (x)")
plt.title("L2 loss, ground truth (y_hat) = 0")
plt.plot(x, L2(x, np.zeros(x.shape)))

x = np.arange(0., 1., 0.01)
plt.subplot(1, 3, 3)
plt.xlabel("Value of input (x)")
plt.title("Cross entropy loss, ground truth (y_hat) = 1")
plt.plot(x, CrossEntropy(x, np.ones(x.shape)))
plt.savefig("loss-func.pdf")
```
